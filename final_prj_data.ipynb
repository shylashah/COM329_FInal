```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```python
coordinates_df = pd.read_csv('world_country_and_usa_states_latitude_and_longitude_values.csv')
netflix_df = pd.read_csv('netflix1.csv')
```

```python
merged = pd.merge(netflix_df, coordinates_df, on='country', how='left')
merged.drop('usa_state_code', axis=1, inplace=True)
merged.drop('usa_state_latitude', axis=1, inplace=True)
merged.drop('usa_state_longitude', axis=1, inplace=True)
merged.drop('usa_state', axis=1, inplace=True)
merged
```

```python
merged.isna().sum()
```

```python
merged['latitude'] = pd.to_numeric(merged['latitude'], errors='coerce')
merged['longitude'] = pd.to_numeric(merged['longitude'], errors='coerce')
merged.dropna(subset=['latitude', 'longitude', 'country_code'], inplace=True)
```

```python
merged.to_csv('netflix_final.csv', index=False)
```
